```python
# 1. Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visual style
sns.set(style="whitegrid")

# 2. Load the dataset
# Update the filename if necessary
df = pd.read_excel('marketing_campaign_dataset.xlsx') 

# 3. Data Cleaning
# Display first few rows and check data info
print(df.head())
print(df.info())

# Drop duplicates or handle missing values if any
df.dropna(inplace=True)

# 4. [span_1](start_span)Exploratory Data Analysis (EDA)
# Example: Calculate ROI (if not already present)
# Formula: (Revenue - Cost) / Cost[span_1](end_span)
if 'ROI' not in df.columns:
    df['ROI'] = (df['Revenue'] - df['Cost']) / df['Cost']

# 5. Visualizing Marketing Channels
# Count of campaigns per channel
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Channel_Used')
plt.title('Distribution of Marketing Channels')
plt.show()

# ROI by Channel
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Channel_Used', y='ROI')
plt.title('Average ROI by Marketing Channel')
plt.show()

# 6. Summary Report Logic
# Group by channel to find the best performing ones
channel_performance = df.groupby('Channel_Used')[['ROI', 'Conversion_Rate']].mean()
print("Performance Summary by Channel:")
print(channel_performance)
```
